```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,5,1)
y1 = x
y2 = x + 1
y3 = x + 2
y4 = x + 3
```

```python
%matplotlib inline
from matplotlib.pyplot import rc # rc는 Run Configration의 약자. 영어가 아닌 언어들이 안 깨지게 출력되도록 해줌.

rc('font',family='New Gulim')


plt.plot(x, y1, '>-m', x, y2,'s:y', x, y3,'p-k', x, y4,'D--b') # 'm'은 마젠타, 'y'는 옐로우 등 색상을 입혀줌.
plt.legend(['북', '쿡', '버', '니'], loc='center')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Saving A Figure')
plt.grid(True)
plt.show()

# 그래프를 이미지 파일로 저장. dpi는 200으로 설정
plt.savefig('saveFigTest3.png', dpi = 200)
plt.show()
```

```python
import matplotlib.pyplot as plt

height = [165, 177, 160, 180, 185, 155, 172]   # 키 데이터
weight = [62, 67, 55, 74, 90, 43, 64]   # 몸무게 데이터


plt.xlabel('Height(Cm)') # x의 라벨 붙이기
plt.ylabel('Weight(Kg)') # y의 라벨 붙이기
plt.title('Height & Weight') # 그래프의 제목

colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 컬러 입히기. 데이터의 개수와 색의 개수가 일치해야 작동함
plt.grid(True)
plt.scatter(height, weight, s = 200, c = colors) # 산점도 (회귀를 하기 전 단계)
```

```python
plt.scatter(height, weight, s=500, c='r') # 마커 크기는 500, 컬러는 붉은색(red)
plt.show()
```

```python
city = ['서울', '인천', '대전', '대구', '울산', '부산', '광주']

# 위도(latitude)와 경도(longitude)
lat  = [37.56, 37.45, 36.35, 35.87, 35.53, 35.18, 35.16] 
lon = [126.97, 126.70, 127.38, 128.60, 129.31, 129.07, 126.85]

# 인구 밀도(명/km^2): 2017년 통계청 자료
pop_den = [16154, 2751, 2839, 2790, 1099, 4454, 2995]

size = np.array(pop_den)*0.2 # 마커의 크기 지정
colors = ['r', 'g', 'b', 'c', 'm', 'k', 'y'] # 마커의 컬러 지정
plt.scatter(lon, lat, s=size, c=colors, alpha=0.1)  # alpha는 투명도

for  i, j, k in zip(lon, lat, city):
    plt.text(i, j, k) # text는 글자 찍어주는 타이틀
```

```python
member_IDs = ['m_01', 'm_02', 'm_03', 'm_04'] # 회원 ID
before_ex = [27, 35, 40, 33] # 운동 시작 전
after_ex = [30,u 38, 42, 37] # 운동 한 달 후
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import rc
rc('font', family = 'New Gulim')

n_data = len(member_IDs)
colors= ['r', 'g', 'b', 'm']
index = np.arange(n_data)

plt.plot(index, before_ex)
plt.bar(index, before_ex, tick_label = member_IDs, color = 'skyblue', width = 0.4, align = 'edge')


plt.bar(index+0.4, after_ex, tick_label = member_IDs, color = 'pink' , width = 0.4, align = 'edge')


plt.legend("회원ID", "윗몸 일으키키 횟수")
plt.xlabel("회원ID")
plt.ylabel("윗몸 일으키기 횟수")
plt.title("운동 전후")
plt.show()

print(index)
```
